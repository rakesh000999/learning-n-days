```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Box Plot
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
```

```python
sns.boxplot(data=data)
plt.title("Box Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
```

```python
# Heatmap
```

```python
data = np.random.randn(10, 10)
```

```python
sns.heatmap(data, annot=True)
plt.title("Heatmap")
plt.show()
```
