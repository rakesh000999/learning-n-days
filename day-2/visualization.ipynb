```python
import matplotlib.pyplot  as plt
```

```python
# Line Chart
x = [1,2,3,4,5,6, 3]
y = [10, 12, 3, 16, 18, 20, 2]
```

```python
plt.plot(x,y)
plt.title('Line Chart')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.show()
```

```python
# Bar chart
```

```python
categories = ['A', 'B', 'C', 'D']
values = [20,3,42,30]
```

```python
plt.bar(categories, values)
plt.title('Title')
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
# Pie Chart
```

```python
labels = ['Apple', 'Mango', 'Avocado']
size = [2,5,3]
```

```python
plt.pie(size, labels = labels, autopct='%1.1f%%', explode= [.2, 0, 0] )
plt.title('Fruits')
plt.legend(title = 'Fruits')
```

```python
# Histogram

import numpy as np
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='r')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
```

```python
# Scatter Plot
```

```python
x = np.random.randn(50)
y = 2 + x + 1 * np.random.randn(50)
```

```python
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
